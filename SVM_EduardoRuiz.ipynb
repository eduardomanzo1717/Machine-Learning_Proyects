# **Support Vector Machine**

El **support vector machine** es una generalización de un clasificador sencillo e intuitivo denominado clasificador de margen máximo. Lamentablemente, este clasificador no se puede aplicar a la mayoría de los conjuntos de datos, ya que requiere que las clases sean separables por un límite lineal.

Se presentara el clasificador de vectores de soporte, una extensión del clasificador de margen máximo que se puede aplicar en una gama más amplia de casos.

Presentaremos el support vector machine, que es una extensión adicional del clasificador de vectores de soporte para adaptarse a los límites de clase no lineales. Los support vector machine están pensados para la clasificación binaria en la que hay dos clases.

Y analizando las extensiones de los support vector machine al caso de más de dos clases.

* *Clasificador de margen máximo*: válido solo si las clases son linealmente separables.

* *Clasificador de vectores de soporte*: admite errores controlados, aplicable a más conjuntos de datos.

* *Support Vector Machine*: incorpora kernels para límites no lineales, ampliando aún más su alcance.

* *Extensiones multiclase*: adaptaciones que permiten aplicar SVM en problemas con más de dos categorías.

**¿Qué es un hiperplano?**
___
En un espacio $p$-dimensional, un *hiperplano* es un subespacio plano afín (es decir, el subespacio no necesita pasar por el origen) de dimensión $p − 1$. Por ejemplo, en dos dimensiones, un hiperplano es un subespacio plano unidimensional. En tres dimensiones, un hiperplano es un subespacio plano bidimensional. En $p > 3$ dimensiones, puede ser difícil visualizar un hiperplano, pero el concepto de subespacio plano $(p − 1)$-dimensional sigue siendo válido. La definición matemática de un hiperplano es bastante sencilla. En dos dimensiones, un hiperplano se define mediante la ecuación

$β_0 + β_1X_1 + β_2X_2 = 0$

para los parámetros $β_0$, $β_1$ y $β_2$.

Extendido al entorno $p$-dimensional:

$β_0 + β_1X_1 + β_2X_2 + ··· + β_pX_p = 0$

**Clasificación mediante un hiperplano separador**
____
Ahora supongamos que tenemos una matriz de datos $X$ de $n × p$ que consta de $n$ observaciones de entrenamiento en un espacio p-dimensional,

$$x_1=
\left(\begin{array}
\\x_{11} \\
\vdots
\\x_{1p}
\end{array}\right),...,
x_n=
\left(\begin{array}
\\x_{n1} \\
\vdots
\\x_{np}
\end{array}\right)
$$

y que estas observaciones se dividen en dos clases: $y_1,...,y_n \in
\{−1, 1\}$, donde $−1$ representa una clase y $1$ la otra.

Supongamos que es posible construir un hiperplano que separe perfectamente las observaciones de entrenamiento según sus etiquetas de clase. En el panel izquierdo de la Figura 2 se muestran ejemplos de tres de estos *hiperplanos de separación*. Podemos etiquetar las observaciones de la clase azul como $y_i = 1$ y las de la clase púrpura como $y_i = −1$. Entonces, un hiperplano de separación tiene la propiedad de que

$β_0 + β1x_{i1} + β_2x_{i2} + ··· + β_px_{ip} > 0$ si $y_i = 1$

Y

$β_0 + β{1}x_{i1} + β_2x_{i2} + ··· + β_px_{ip} < 0$ si $y_i = −1$.

De manera equivalente, un hiperplano separador tiene la propiedad de que

$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) > 0$

para todo $i = 1,...,n$.

Si existe un hiperplano separador, podemos usarlo para construir un clasificador muy natural: a una observación de prueba se le asigna una clase según su lado del hiperplano. El panel derecho de la Figura 2 muestra un ejemplo de dicho clasificador. Es decir, clasificamos la observación de prueba $x^*$ según el signo de $f(x^*) = β_0+β_1x^*_1+β_2x^*_2+···+β_px^*_p$. Si $f(x^*)$ es positiva, asignamos la observación de prueba a la clase $1$, y si $f(x^*)$ es negativa, la asignamos a la clase $−1$. También podemos usar la magnitud de $f(x^*)$.

Tres hiperplanos de separación, de entre muchos posibles, se muestran en negro. Derecha: Un hiperplano de separación se muestra en negro. La cuadrícula azul y morada indica la regla de decisión que toma un clasificador con base en este hiperplano de separación: una observación de prueba que se encuentre en la parte azul de la cuadrícula se asignará a la clase azul, y una observación de prueba que se encuentre en la parte morada se asignará a la clase morada.

**El clasificador de margen máximo**
___
Una elección natural es el *hiperplano de margen máximo* (también conocido como *hiperplano de separación óptimo*), que es el hiperplano de separación que está más alejado de las observaciones de entrenamiento.

Si $β_0, β_1,..., β_p$ son los coeficientes del hiperplano de margen máximo, entonces el clasificador de margen máximo clasifica la observación de prueba $x^*$ según el signo de $f(x^*) = β_0 + β_1x^*_1 + β_2x^*_2 + ··· + β_px^*_p$.

Al examinar la Figura 3, observamos que tres observaciones de entrenamiento son equidistantes del hiperplano de margen máximo y se encuentran a lo largo de las líneas discontinuas que indican la anchura del margen. Estas tres observaciones se conocen como *vectores de soporte*, ya que son vectores en un espacio p-dimensional (en la Figura 3, p = 2) y "soportan" el hiperplano de margen máximo, en el sentido de que si estos puntos se movieran ligeramente, el hiperplano de margen máximo también se movería.

El hiperplano de margen máximo depende directamente de los vectores de soporte.

image.png

**Figura 3**.- Existen dos clases de observaciones, mostradas en azul y en morado. El hiperplano de margen máximo se muestra como una línea continua. El margen es la distancia desde la línea continua hasta cualquiera de las líneas discontinuas. Los dos puntos azules y el punto morado que se encuentran sobre las líneas discontinuas son los vectores de soporte, y la distancia desde estos puntos hasta el hiperplano se indica mediante flechas. La cuadrícula morada y azul indica la regla de decisión tomada por un clasificador con base en este hiperplano separador.

**Construcción del clasificador de margen máximo**
___
Ahora consideramos la tarea de construir el hiperplano de margen máximo basado en un conjunto de $n$ observaciones de entrenamiento $x_1,...,x_n \in R_p$ y etiquetas de clase asociadas $y_1,...,y_n \in \{−1, 1\}$. En resumen, el hiperplano de margen máximo es la solución al problema de optimización

$$
\text{maximize}_{\beta_0, \beta_1, \ldots, \beta_p, M}\; M
$$

$$
\text{sujeto a} \sum_{j=1}^p \beta_j^2=1
$$

$$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) ≥ M  \; ∀i=1,...,n$$

La restricción $$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) ≥ M  \; ∀i=1,...,n$$

garantiza que cada observación estará en el lado correcto del hiperplano, siempre que $M$ sea positivo.

Obsérvese que $$
 \sum_{j=1}^p \beta_j^2=1
$$ no es realmente una restricción del hiperplano, ya que si $$β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip} = 0$$ define un hiperplano, entonces también lo define $$k(β_0 + β_1x_{i1} + β_2x_{i2} + ···+β_px_{ip})=0$$ para cualquier $k \neq 0 $. Con esta restricción la distancia perpendicular desde la i-ésima observación al hiperplano esta dada por $y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip})$.

Por lo tanto, las restricciones $$
\sum_{j=1}^p \beta_j^2=1
$$

$$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) ≥ M  \; ∀i=1,...,n$$

garantizan que cada observación esté en el lado correcto del hiperplano y al menos a una distancia $M$ de este. Por lo tanto, $M$ representa el margen de nuestro hiperplano, y el problema de optimización elige $β_0, β_1,..., β_p$ para maximizar $M$. *¡Esta es exactamente la definición del hiperplano de margen máximo!*

**Definiciones extra:**
___
  Supongamos que queremos estimar una función verdadera $f(x)$ usando un estimador $\hat f(x)$, obtenido a partir de un conjunto de entrenamiento.
  
  **Sesgo** (bias):
  Mide el error sistemático del modelo, es decir, la diferencia entre la predicción promedio del estimador y el valor verdadero:
  $$
  Sesgo(x)=E[\hat{f}(x)]-f(x).
  $$
  Un modelo con alto sesgo simplifica demasiado el problema y tiende a cometer errores sistemáticos.

  **Varianza** (variance):
  Mide cuánto varían las predicciones del modelo si entrenamos con diferentes muestras de datos. Formalmente:
  $$
  Varianza(x)=E[(\hat{f}(x)−E[\hat{f}(x)])^2].
  $$
  Un modelo con alta varianza es muy sensible a las fluctuaciones en los datos de entrenamiento.

**Detalles del clasificador de vectores de soporte**
___
El *clasificador de vectores de soporte* es una extensión del clasificador de margen máximo. Su objetivo es encontrar un hiperplano que separe lo mejor posible las observaciones de entrenamiento en dos clases, permitiendo cierta tolerancia a errores.

Considerrando la solución al problema de optimización

$$
\text{maximize}_{\beta_0, \beta_1, \ldots, \beta_p, \epsilon_1, ..., \epsilon_n, M}\; M
$$

$$
\text{sujeto a} \sum_{j=1}^p \beta_j^2=1
$$

$$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) ≥ M(1-\epsilon_i),  \; ∀i=1,...,n$$

$$
\epsilon_i \geq 0, \; \sum_{i=1}^n \epsilon_i\leq C
$$

donde $C$ es un parámetro de ajuste no negativo. Al igual que en el caso del margen máximo, buscamos maximizar la anchura del margen $M$. Sin embargo, ahora introducimos las variables de holgura $\epsilon_i$​, que permiten que algunas observaciones estén en el lado incorrecto del margen o incluso del hiperplano.

* Si $\epsilon_i=0$, la observación está correctamente clasificada y fuera del margen.

* Si $\epsilon_i >0$, la observación viola el margen.

* Si $\epsilon_i>1$, la observación se encuentra en el lado incorrecto del hiperplano.

El número y la magnitud de estas violaciones está controlado por el parámetro de ajuste $C$. Este funciona como un “presupuesto” de errores:

* Cuando $C=0$, no se permiten violaciones del margen; el problema se reduce al clasificador de margen máximo, que solo existe si las clases son perfectamente separables.

* Cuando $C>0$, se permite cierto grado de error. Cuanto mayor sea $C$, más observaciones pueden violar el margen o incluso quedar en el lado incorrecto del hiperplano.

Esto implica un equilibrio entre *sesgo* y *varianza*:

* Con un valor grande de $C$, el clasificador permite muchos vectores de soporte, ajustándose de forma más flexible a los datos. El resultado suele tener menor varianza, pero puede presentar mayor sesgo.

* Con un valor pequeño de $C$, el clasificador utiliza pocos vectores de soporte, lo que lo hace más ajustado a los datos de entrenamiento. En este caso, obtenemos bajo sesgo, pero alta varianza.

Una propiedad esencial del clasificador de vectores de soporte es que únicamente las observaciones que se encuentran en el margen, o que lo violan, influyen en la posición del hiperplano. Dichas observaciones reciben el nombre de vectores de soporte y constituyen el subconjunto de datos del que realmente depende el clasificador. En cambio, las observaciones que están correctamente clasificadas y situadas fuera del margen no ejercen ningún efecto sobre la frontera de decisión.

Gracias a esto, el clasificador de vectores de soporte es robusto frente a observaciones alejadas del límite de decisión, lo que lo diferencia de otros métodos como el análisis discriminante lineal, donde todas las observaciones contribuyen al cálculo de la regla de clasificación.

En resumen, el clasificador de vectores de soporte encuentra un hiperplano que separa lo mejor posible las clases, controlando la tolerancia a errores mediante el parámetro $C$, y apoyándose únicamente en los vectores de soporte para definir la frontera de decisión.

**Clasificación con límites de decisión no lineales**
___
Los Support vector machine funcionan bien cuando el límite entre clases es lineal, pero su desempeño cae si el límite es no lineal. Para mejorar esto, al igual que en regresión lineal con relaciones no lineales, se puede ampliar el espacio de características usando funciones polinómicas (cuadráticas, cúbicas o de mayor orden) de los predictores.

Por ejemplo, en lugar de ajustar un clasificador de vectores de soporte utilizando $p$ características

$$X_1, X_2,...,X_p$$

Podríamos ajustar un clasificador de vectores de soporte utilizando $2p$ características

 $$X_1, X^2_1, X_2, X_2^2,...,X_p, X^2_p$$

Entonces

$$
\text{maximize}_{\beta_0, \beta_1, \ldots, \beta_p, \epsilon_1, ..., \epsilon_n, M}\; M
$$

$$
\text{sujeto a} \sum_{j=1}^p \beta_j^2=1
$$

$$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) ≥ M(1-\epsilon_i),  \; ∀i=1,...,n$$

$$
\epsilon_i \geq 0, \; \sum_{i=1}^n \epsilon_i\leq C
$$

Se convierte en

$$
\text{maximize}_{\beta_0, \beta_{11}, \beta_{12},..., \beta_{p1}, \beta_{p2}, \epsilon_1, ..., \epsilon_n, M}\; M \; (**)
$$

$$
\text{sujeto a } y_i(β_0 + \sum_{j=1}^p \beta_{j1}x_{ij}+\sum_{j=1}^p \beta_{j2}x^2_{ij}) ≥ M(1-\epsilon_i),  \; ∀i=1,...,n
$$

$$
\epsilon_i \geq 0, \; \sum_{i=1}^n \epsilon_i\leq C, \; \sum_{j=1}^p \sum_{k=1}^2\beta_{jk}^2=1
$$

¿Por qué esto conduce a un límite de decisión no lineal? En el espacio de características ampliado, el límite de decisión que resulta de $(**)$ es lineal. Sin embargo, en el espacio original, el límite de decisión tiene la forma $q(x)=0$, donde $q$ es un polinomio cuadrático y sus soluciones suelen ser no lineales. Se pueden incluir términos polinómicos de orden superior o términos de interacción $XjXj'$ para $j\neq j'$, o incluso otras funciones de los predictores. Aunque hay muchas formas de ampliar el espacio de características, un aumento excesivo puede hacer que los cálculos sean inmanejables. El support vector machine permite ampliar el espacio de características de manera que los cálculos sigan siendo eficientes.

**El Support Vector Machine**
___
El *Support Vector Machine* es una extensión del clasificador de vectores de soporte que resulta de ampliar el espacio de características de una manera específica mediante kernels. El enfoque de *kernel* que describimos aquí es simplemente un método computacional eficiente para implementar esta idea.

Hasta ahora noo se ha explicado cómo se calcula el clasificador de vectores de soporte debido a que los detalles se vuelven algo técnicos. Sin embargo, resulta que la solución al problema del clasificador de vectores de soporte

$$
\text{maximize}_{\beta_0, \beta_1, \ldots, \beta_p, \epsilon_1, ..., \epsilon_n, M}\; M
$$

$$
\text{sujeto a} \sum_{j=1}^p \beta_j^2=1
$$

$$y_i(β_0 + β_1x_{i1} + β_2x_{i2} + ··· + β_px_{ip}) ≥ M(1-\epsilon_i),  \; ∀i=1,...,n$$

$$
\epsilon_i \geq 0, \; \sum_{i=1}^n \epsilon_i\leq C
$$


involucra solo los productos internos de las observaciones (a diferencia de las observaciones mismas). El producto interno de dos $r$-vectores $a$ y $b$ se define como

$$\langle a, b\rangle = \sum_{i=1}^r a_i b_i$$

Por lo tanto, el producto interno de dos observaciones $x_{i}, x_{i'}$ está dado por

$$\langle x_i, x_{i'}\rangle = \sum_{j=1}^p x_{ij} x_{i'j}$$

Se puede demostrar que



*   El clasificador de vectores de soporte lineal se puede representar como $$f(x)=\beta_0+\sum_{i=1}^n \alpha_i\langle x,x_i \rangle $$ donde hay $n$ parámetros $α_i, \; i = 1,...,n$, uno por cada observación de entrenamiento.
*   Para estimar los parámetros $α_1,..., α_n$ y $β_0$, todo lo que necesitamos son los $\left(\begin{array}\\n \\2\end{array}\right)$ productos internos $\langle x_i, x_{i'}\rangle$ entre todos los pares de observaciones de entrenamiento (La notación $\left(\begin{array}\\n \\2\end{array}\right)=\frac{n(n − 1)}{2}$).

Para evaluar la función $f(x)$, necesitamos calcular el producto interno entre el nuevo punto $x$ y cada uno de los puntos de entrenamiento $x_i$. Sin embargo, resulta que $α_i$ es distinto de cero solo para los vectores de soporte en la solución, es decir, si una observación de entrenamiento no es un vector de soporte, entonces su $α_i$ es igual a cero. Por lo tanto, si $S$ es el conjunto de índices de estos puntos de soporte, podemos reescribir cualquier función solución como

$$f(x)=\beta_0+\sum_{i\in S} \alpha_i\langle x,x_i \rangle $$

la cual tiene menos términos que la forma anterior.

Supongamos ahora que cada vez que el producto interno $$\langle x_i, x_{i'}\rangle = \sum_{j=1}^p x_{ij} x_{i'j}$$ aparece en la representación $$f(x)=\beta_0+\sum_{i=1}^n \alpha_i\langle x,x_i \rangle $$ ó en el cálculo de la solución del clasificador de vectores de soporte, lo reemplazamos por una generalización del producto interno de la forma $$K(x_i,x_{i'})$$ donde K es una función a la que llamaremos *kernel*. Un *kernel* es una función que cuantifica la similitud de dos observaciones. Por ejemplo, podríamos simplemente tomar $$K(x_i,x_{i'})=\sum_{j=1}^p x_{ij} x_{i'j} $$ que se conoce como kernel lineal porque el clasificador de vectores de soporte es lineal en sus características. También se podría reemplazar cada instancia de $$\sum_{j=1}^p x_{ij} x_{i'j}$$ con la cantidad
$$K(x_i,x_{i'})=(1+\sum_{j=1}^p x_{ij} x_{i'j})^d $$ Esto se conoce como núcleo polinomial de grado d, donde d es un entero positivo. El uso de dicho núcleo con d > 1, en lugar del núcleo lineal estándar, en el algoritmo del clasificador de vectores de soporte, genera una frontera de decisión mucho más flexible.



Nótese que en este caso la función (no lineal) tiene la forma $$f(x)=\beta_0+\sum_{i\in S} \alpha_iK(x,x_i) $$

image.png

**Figura 9**.- Izquierda: Se aplica una SVM con un núcleo polinomial de grado 3 a los datos no lineales de la Figura 8, lo que resulta en una regla de decisión mucho más apropiada. Derecha: Se aplica una SVM con un núcleo radial. En este ejemplo, cualquiera de los dos núcleos es capaz de capturar la frontera de decisión.

El panel izquierdo de la Figura 9 muestra un ejemplo de una SVM con un núcleo polinomial aplicado a los datos no lineales de la Figura 8. El ajuste supone una mejora sustancial respecto al clasificador de vectores de soporte lineal. Cuando $d = 1$, el Support vector machine se reduce al clasificador de vectores de soporte. El kernel polinomial mostrado es un ejemplo de un posible kernel no lineal, pero existen numerosas alternativas.

Otra opción es el kernel radial, que adopta la forma $$K(x_i,x_{i'})=exp(-\gamma\sum_{j=1}^p (x_{ij} -x_{i'j})^2) $$

$\gamma$ es una constante positiva.

¿Cómo funciona realmente el kernel radial? Si una observación de prueba dada $x^* = (x^*_1,...,x^*_p)^T$ está lejos de una observación de entrenamiento $x_i$ en términos de distancia euclidiana, entonces $$\sum_{j=1}^p (x^*_{j} -x_{ij})^2$$ será grande, y por lo tanto, $$K(x^*,x_{i})=exp(-\gamma\sum_{j=1}^p (x^*_{j}- x_{ij})^2) $$ será pequeño. Esto significa que $x_i$ prácticamente no desempeñará ningún papel en $f(x^*)$. Las observaciones de entrenamiento alejadas de $x^*$ prácticamente no influirán en la etiqueta de clase predicha para $x^*$. Esto significa que el kernel radial tiene un comportamiento muy local.

¿Cuál es la ventaja de usar un kernel en lugar de simplemente ampliar el espacio de características utilizando funciones de las características originales? Una ventaja es computacional, y consiste en que al usar kernels, solo se necesita calcular $K(x_i, x_{i'})$ para todos los $\left(\begin{array}\\n \\2\end{array}\right)$ pares distintos $i$, $i'$. En muchas aplicaciones de SVM, este espacio de características ampliado es tan grande que los cálculos son inviables.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import make_blobs
```

```python
# Generar algunos datos para representar "celulares" y "tablets" con características de largo y ancho en cm
# Ajustamos los centros para representar la idea de que los celulares tienden a ser más pequeños y los tablets más grandes
X, y = make_blobs(n_samples=100, centers=[[5, 10], [15, 20]], cluster_std=[2, 3], random_state=6)

# Asignar etiquetas conceptuales a las clases
y_labels = np.where(y == 0, 'Celulares', 'Tablets')
```

```python
# Crear un clasificador SVM con un kernel lineal
# C es el parámetro de penalización del término de error.
# Controla el equilibrio entre un límite de decisión suave y la clasificación correcta de los puntos de entrenamiento.
clf = svm.SVC(kernel='linear', C=1000)
clf.fit(X, y)
```

```python
# Graficar los puntos de datos
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=y, s=30, cmap=plt.cm.Paired)

# Graficar la función de decisión
ax = plt.gca()
xlim = ax.get_xlim()
ylim = ax.get_ylim()

# Crear una cuadrícula para evaluar el modelo
xx = np.linspace(xlim[0], xlim[1], 30)
yy = np.linspace(ylim[0], ylim[1], 30)
YY, XX = np.meshgrid(yy, xx)
xy = np.vstack([XX.ravel(), YY.ravel()]).T
Z = clf.decision_function(xy).reshape(XX.shape)

# Graficar el límite de decisión y los márgenes
ax.contour(XX, YY, Z, colors='k', levels=[-1, 0, 1], alpha=0.5,
           linestyles=['--', '-', '--'])

# Graficar los vectores de soporte
ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
           facecolors='none', edgecolors='k')
plt.title('SVM con Kernel Lineal (Celulares vs Tablets)')
plt.xlabel('Ancho (cm)')
plt.ylabel('Largo (cm)')
plt.show()
```

**Ejemplo** con Kernel Lineal (Relacionado con el Clasificador de Margen Máximo y el Clasificador de Vectores de Soporte)

Utiliza un kernel lineal, se conecta directamente con los conceptos del **Clasificador de Margen Máximo** y el **Clasificador de Vectores de Soporte**.

*   **La Idea Principal:** El objetivo es encontrar una línea (un **hiperplano** en dos dimensiones) que separe las dos clases de datos de entrenamiento.
*   **Buscando el Mejor Hiperplano:** Entre todas las líneas que pueden separar las clases, el **Clasificador de Margen Máximo** elige aquella que está más lejos de la observación de entrenamiento más cercana de cualquiera de las clases. Esa distancia máxima se llama **margen**. El gráfico con el kernel lineal muestra esta línea de decisión y las líneas discontinuas que definen el margen.
*   **Los Vectores de Soporte:** Como se explica, solo unas pocas observaciones de entrenamiento, llamadas **vectores de soporte**, son las que realmente determinan la posición del hiperplano de margen máximo. Estos son los puntos que se encuentran justo en el margen. En el gráfico, los círculos alrededor de algunos puntos señalan estos vectores de soporte. Si movieras cualquier otro punto que no sea un vector de soporte, el hiperplano no cambiaría.
*   **¿Qué pasa si las Clases no se separan perfectamente? (El Clasificador de Vectores de Soporte y el parámetro C):** En la vida real los datos rara vez son perfectamente separables por una línea. Aquí es donde entra el **Clasificador de Vectores de Soporte**. Permite que algunas observaciones estén en el lado incorrecto del margen, o incluso del hiperplano, introduciendo un poco de flexibilidad. El parámetro $C$ en el código controla cuánto "presupuesto" de error permitimos. Un $C$ grande penaliza mucho los errores, buscando un margen estrecho pero con pocos errores en los datos de entrenamiento (similar al margen máximo para datos separables). Un $C$ más pequeño permite más errores para conseguir un límite de decisión más suave y con mayor capacidad de generalización.

**Ejemplo** con Kernel No Lineal

Usando el kernel, ilustra la extensión más poderosa, el **Support Vector Machine** propiamente dicho, que permite encontrar límites de decisión no lineales.

*   **El Desafío de los Límites No Lineales:** A menudo, las clases en los datos no pueden separarse eficazmente con una línea recta. Se presentan ejemplos como la Figura 8 donde el límite entre clases es curvo.
*   **El Truco de los Kernels:** En lugar de intentar encontrar una línea en el espacio original, el SVM con kernels no lineales utiliza un truco matemático. Piensen en ello como si el SVM "mapeara" o transformara los datos a un espacio de dimensiones mucho más altas donde sí es posible encontrar un hiperplano lineal que separe las clases. La magia es que no necesitamos calcular explícitamente ese espacio de alta dimensión. El **kernel** es una función $$K(x_i,x_{i'})$$ que calcula la similitud entre dos puntos de datos como si ya estuvieran en ese espacio de alta dimensión.
*   **El Kernel Radial:** El ejemplo usa el kernel radial, $$K(x_i,x_{i'})=exp(-\gamma\sum_{j=1}^p (x_{ij} -x_{i'j})^2) $$ Este kernel, como se describe anteriormente, hace que las observaciones cercanas en el espacio original tengan una gran influencia en la clasificación de un nuevo punto, mientras que las observaciones lejanas tienen poca influencia. Esto permite que el límite de decisión se curve para adaptarse a la forma de los datos.
*   **El Límite Curvo:** El gráfico del ejemplo con kernel muestra claramente un límite de decisión que no es una línea recta, sino una curva que se adapta mejor a la distribución de los datos no lineales. Este límite curvo en nuestro espacio original corresponde a un hiperplano lineal en ese espacio "virtual" de alta dimensión que crea el kernel.
*   **Los Vectores de Soporte en el Caso No Lineal:** También aquí, los puntos marcados con círculos son los vectores de soporte. Estos son los puntos de entrenamiento que, en el espacio de alta dimensión, están en el margen o dentro del margen del hiperplano lineal. Son estos puntos los que definen la forma del límite de decisión curvo en nuestro espacio original.
*   **El Parámetro Gamma:** El parámetro `gamma` controla la "amplitud" de la influencia de cada vector de soporte. Un `gamma` alto significa que solo los puntos muy cercanos importan, lo que puede llevar a un límite de decisión muy irregular. Un `gamma` bajo significa que los puntos más lejanos también tienen influencia, resultando en un límite más suave.


***SVM con más de dos clases***
___

**Clasificación uno contra uno (one-vs-one)**
___

Supongamos que queremos realizar una clasificación con SVM en el caso de que existan $K>2$ clases. Una estrategia común es el enfoque uno contra uno (one-vs-one), en el que se construyen $\left(\begin{array}\\n \\2\end{array}\right)$ clasificadores SVM, cada uno entrenado para distinguir entre un par de clases.

Por ejemplo, un clasificador puede diferenciar entre la clase $k$ (dadas como $+1$) y la clase $k'$ (dadas como $-1$).

Para clasificar una nueva observación, se hace que pase por los $\left(\begin{array}\\n \\2\end{array}\right)$ clasificadores.

Cada uno de ellos asigna la observación a una de las dos clases involucradas en esa comparación. Luego, se cuenta cuántas veces la observación fue asignada a cada clase. Finalmente, la observación se clasifica en la clase con el mayor número de “votos” obtenidos a lo largo de todas las comparaciones por pares.

**Clasificación de uno contra todos (one-vs-all)**
___

El enfoque uno contra todos (también llamado one-vs-all) es otra estrategia para extender los SVM al caso de $K>2$ clases. En este procedimiento se entrenan $K$ clasificadores SVM: cada uno compara una clase específica con el conjunto de las $K-1$ clases restantes.

Sea $\beta_{0k},\beta_{1k},...,\beta_{pk} $ el conjunto de parámetros del clasificador correspondiente a la clase $k$, obtenido al ajustar el SVM que distingue a la clase $k$ (dadas como $+1$) frente a todas las demás (dadas como $-1$).

Dada una observación de prueba $x^*$, evaluamos las funciones de decisión:
$$f_k(x^*)=\beta_{0k}+\beta_{1k}x_1^*+\beta_{2k}x_2^*+...+\beta_{pk}x_p^*$$
Finalmente, asignamos la observación a la clase $k$ que produzca el valor más alto de $f_k(x^*)$, lo cual indica el mayor nivel de confianza de que la observación pertenece a dicha clase frente al resto.

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn import svm
from sklearn.datasets import load_wine
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
# Cargar el conjunto de datos Wine
wine = load_wine()
X = wine.data
y = wine.target
target_names = wine.target_names
```

```python
# Primero, escalamos los datos, lo cual es importante para PCA.
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
# Aplicar PCA para reducir a 2 componentes principales
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)
```

```python
# Crear un clasificador SVM con un kernel lineal
# Usamos el enfoque one-vs-one (por defecto en SVC para multi-clase)
clf_wine = svm.SVC(kernel='linear', C=1000)
clf_wine.fit(X_pca, y)
```

```python
# Graficar los puntos de datos reducidos
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
lw = 2
for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)
```

```python
# Graficar los puntos de datos reducidos
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise', 'darkorange']
lw = 2

for color, i, target_name in zip(colors, [0, 1, 2], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.8, lw=lw,
                label=target_name)

# Graficar las regiones de decisión para el SVM multi-clase en el espacio PCA
ax_wine = plt.gca()
xlim_wine = ax_wine.get_xlim()
ylim_wine = ax_wine.get_ylim()

xx_wine = np.linspace(xlim_wine[0], xlim_wine[1], 100)
yy_wine = np.linspace(ylim_wine[0], ylim_wine[1], 100)
YY_wine, XX_wine = np.meshgrid(yy_wine, xx_wine)
xy_wine = np.vstack([XX_wine.ravel(), YY_wine.ravel()]).T

# Predecir la clase para cada punto en la cuadrícula en el espacio PCA
Z_wine = clf_wine.predict(xy_wine).reshape(XX_wine.shape)

# Graficar las regiones de decisión usando contourf
ax_wine.contourf(XX_wine, YY_wine, Z_wine, cmap=plt.cm.coolwarm, alpha=0.4)

# Graficar los vectores de soporte en el espacio PCA
ax_wine.scatter(clf_wine.support_vectors_[:, 0], clf_wine.support_vectors_[:, 1], s=100,
           facecolors='none', edgecolors='k', linewidths=1.5)


plt.title('SVM con Kernel Lineal en Datos Wine (PCA Reducido)')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.show()
```

***Regresión logística***
___
La *regresión logística* es un método estadístico de clasificación, no de regresión en el sentido clásico.

Su objetivo no es predecir un número continuo, sino la probabilidad de que ocurra un evento.

Por ejemplo:

* ¿Un cliente comprará (sí/no)?

* ¿Un correo es spam (sí/no)?

* ¿Un paciente tiene cierta enfermedad (sí/no)?

1. Idea principal

Usamos las variables predictoras $X_1,X_2,…,X_p$ para estimar la probabilidad de que la variable de salida $Y$ tome el valor $1$.
En lugar de un modelo lineal simple como en regresión lineal, aplicamos una transformación para que la salida siempre esté entre $0$ y $1$ (pues son probabilidades).

2. La función logística

La función que usamos es
$$p(X)=\frac{1}{1+e^{-(\beta_0+\beta_1X_1+...+\beta_pX_p)}} $$
Aquí $p(X)$ es la probabilidad de que $Y=1$.

La curva tiene forma de $S$, asegurando valores entre $0$ y $1$.

3. Interpretación

El modelo no predice directamente "sí" o "no", sino una probabilidad. Con un umbral de decisión:

* Si $p(X)>0.5$, clasificamos como $1$ (evento ocurre).

* Si $p(X)≤0.5$, clasificamos como 0 (evento no ocurre).

4. Ventajas

Produce probabilidades interpretables, no solo etiquetas.

Funciona bien cuando la relación entre las variables y la clase es aproximadamente lineal en los log-odds.

Es más estable que las SVM si hay mucha superposición entre clases.

***Relación entre SVM y Regresión Logística***
___
Cuando las máquinas de soporte vectorial (SVMs) aparecieron a mediados de los 90, llamaron mucho la atención en estadística y aprendizaje automático. Parte de su popularidad se debió a:

* su buen desempeño en la práctica,

* un gran impulso de marketing,

* y la percepción de que ofrecían un enfoque radicalmente distinto a métodos clásicos como la regresión logística o el análisis discriminante lineal.

La idea de buscar un hiperplano que separe los datos lo mejor posible, permitiendo algunas violaciones, parecía novedosa. Además, el uso de kernels para ampliar el espacio de características y manejar fronteras no lineales se veía como una innovación exclusiva de las SVM.

Con el tiempo se descubrió que existen profundas similitudes entre las SVM y los métodos estadísticos clásicos:

* El problema de optimización que define el SVM, puede reescribirse en una forma muy parecida a la regresión logística penalizada.

* En particular, *el clasificador de vectores de soporte lineal* se ajusta resolviendo:

$$
min_{\beta_{0},\beta_{1},...,\beta_{p}} \left\{\sum_{i=1}^n max[0,1-y_if(x_i)]+\lambda\sum_{j=1}^p\beta_j^2\right\}
$$

donde
$$
f(x)=\beta_0+\beta_1x_1+...+\beta_px_p
$$

* El primer término es la función de pérdida hinge (hinge loss).

* El segundo es una penalización tipo ridge que controla el compromiso entre sesgo y varianza.

En forma general:

$$
min_{\beta_{0},\beta_{1},...,\beta_{p}} \left\{L(X,y,\beta)+\lambda P(\beta)\right\}
$$

donde:

* $L(X,y,\beta)$ mide el ajuste del modelo,

* $P(\beta)$ es una penalización sobre los coeficientes.

Ejemplos:

En ridge regression:
$$L(X,y,\beta)=\sum_{i=1}^n\left(y_i-\beta_0-\sum_{j=1}^px_{ij}\beta_j\right)^2,  \; P(\beta)=\sum_{j=1}^p\beta_j^2 $$

En el lasso:
$$P(\beta)=\sum_{j=1}^p|\beta_j|$$

En la SVM:
$$L=\sum_{i=1}^n max[0,1-y_i(\beta_0+\beta_1x_{i1}+...+\beta_px_{ip})] $$

*Comparación con Regresión Logística*

* En SVM, la pérdida hinge es exactamente 0 cuando la observación queda en el lado correcto del margen:
$$y_i(\beta_0+\beta_1x_{i1}+...+\beta_px_{ip})$$

* En regresión logística, la pérdida es:
$$
ℓ(y_i,f(x_i))=log(1+e^{-y_if(x_i)})
$$
que nunca es cero, pero se vuelve muy pequeña cuando la observación está bien clasificada y lejos de la frontera.

Cuando $y_i(\beta_0+\beta_1x_{i1}+...+\beta_px_{ip})$ es mayor que $1$, la pérdida de la SVM es cero, ya que esto corresponde a una observación que se encuentra en el lado correcto del margen. En general, ambas funciones de pérdida tienen un comportamiento bastante similar.
